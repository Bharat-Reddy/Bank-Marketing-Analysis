# AIM : To find out how many Features are required to capture 80% of variance by performing PCA

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
data_ohe = pd.read_csv(r"..\Data\full_ohe.csv")
data_ohe.head()
```

```python
data_le = pd.read_csv(r".\full_le.csv")
data_le.head()
```

## Appliying PCA on Label Encoded data

```python
# raw data
X = data_le.drop('y', axis=1).values
y = data_le['y'].values
```

```python
pca = PCA(n_components=0.8)
x_pca = pca.fit_transform(X)
X.shape, x_pca.shape
```

## Appliying PCA on One Hot Encoded data

```python
# raw data
X = data_ohe.drop('y', axis=1).values
y = data_ohe['y'].values
```

```python
pca = PCA(n_components=0.8)
x_pca = pca.fit_transform(X)
X.shape, x_pca.shape
```

# Observation1 : 
## We need 17 Features out of 55 Features to capture 80% of variance of one hot encoded data

# Observation2 :
## We need 3 Features out of 20 Features to capture 80% of variance of Label encoded data
